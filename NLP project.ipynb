```python
import pandas as pd
data=pd.read_csv("emotion_sentimen_dataset.csv")
```

```python
data.head()
```

```python
df=data.head(5000)
df
```

```python
df['Emotion'].value_counts()
```

```python
df['new']=df['Emotion'].apply(lambda x: 1 if x=='love' else 0)
```

```python
df.head(20)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
ps=PorterStemmer()
lemmatizer=WordNetLemmatizer()
stopword=set(stopwords.words('english'))
```

```python
df['tokens']=df['text'].apply(word_tokenize)
```

```python
df.head()
```

```python
df['tokens'] = df['tokens'].apply(lambda x: x if isinstance(x, list) else [])
```

```python
df.head()
```

```python
df['stopwords']=df['tokens'].apply(lambda x:[word.lower() for word in x if word.lower() not in stopword and word.isalpha()])
```

```python
df
```

```python
df['stemming']=df['stopwords'].apply(lambda x:[ps.stem(word) for word in x])
```

```python
df
```

```python
df['lemma']=df['stopwords'].apply(lambda x:[lemmatizer.lemmatize(word) for word in x])
```

```python
df.head()
```

```python
df['processed_text']=df['lemma'].apply(lambda x:" ".join(x))  #corrected to join the actual lemmas
```

```python
Vectorizer=CountVectorizer()
x=Vectorizer.fit_transform(df['processed_text'])
```

```python
df.head()
```

```python
words=Vectorizer.get_feature_names_out()
```

```python
words
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(bow,df['new'],test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
